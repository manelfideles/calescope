```python
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
```

```python
def encodeLocationNames(df: pd.DataFrame, locationIds=locationIds):
	droga = df.columns.to_list()
	for location in locationIds.items():
		try:
			idx = droga.index(location[1])
		except ValueError:
			idx = -1
		if idx >= 0:
			droga[idx] = location[0]
	return droga
```

```python
dataset2020.columns = encodeLocationNames(dataset2020)
dataset2020 = dataset2020.rename(columns={'Altitude_sonda': 'height'})
```
